# Load ED 2015 SAS data

Unzip `ed2015-sas.sas7bdat.zip` and load the `.sas7bdat` file into a pandas DataFrame.

```python
import zipfile
from pathlib import Path

import pandas as pd
```

```python
data_dir = Path(".").resolve()  # notebook directory when run from med_proj/data
zip_path = data_dir / "ed2015-sas.sas7bdat.zip"

with zipfile.ZipFile(zip_path, "r") as zf:
    zf.extractall(path=data_dir)

sas_path = data_dir / "ed2015-sas.sas7bdat"
df = pd.read_sas(sas_path)
```
